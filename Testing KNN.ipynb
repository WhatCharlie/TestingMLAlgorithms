```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Define the number of rows and columns
num_rows = 23
num_columns = 2

# Create a DataFrame with random data
random_data = pd.DataFrame(np.random.rand(num_rows, num_columns), 
                            columns=[f'Column_{i+1}' for i in range(num_columns)])

print(random_data)

rd = random_data
rd['Column_1'] = (rd['Column_1'] * num_rows)
rd['Column_2'] = (rd['Column_2'] * num_rows)

print(random_data)
```

```python
def round_to_odd(number):
    rounded = round(number)
    if rounded % 2 == 0:
        return rounded + 1 if number > rounded else rounded - 1
    return rounded
print(f"K = {round_to_odd(np.sqrt(num_rows*num_columns))}")

K = round_to_odd(np.sqrt(num_rows*num_columns))


def find_K_lowest_distances(xyz_list, K_int, target_point):
    k_counter = 0
    xyz_copy = xyz_list
    closest_points = []
    c1_probability = 0.0
    c2_probability = 0.0
    while k_counter < K:
        for sublist in xyz_copy:
                if sublist[2] == min(xyz_copy, key=lambda x: x[2])[2]:
                    closest_points.append(sublist)
                    xyz_copy.remove(sublist)
                    k_counter+=1
    for sublist in closest_points:
        if sublist[3] == c1:
            c1_probability += 1
        elif sublist[3] == c2:
            c2_probability += 1
    c1_probability = c1_probability / K_int
    c2_probability = c2_probability / K_int
    if c1_probability > 0.5:
        '''print(f"Probability that Black Dot is Blue = {c1_probability * 100}%")'''
        return Blue
    elif c1_probability < 0.5:
        '''print(f"Probability that Black Dot is Orange = {c2_probability * 100}%")'''
        return Orange
    else: 
        '''print("Not enough information to determine the New Dot's Class")'''
        return Grey
black_data_coords_list = make_Coordinate_List(black_x, black_y)
print(find_K_lowest_distances(black_data_coords_list, K, black_Dot))
```
